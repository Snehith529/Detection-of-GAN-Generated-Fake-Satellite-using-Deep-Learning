```python
from PIL import Image
import os
import random
import pandas as pd
import glob
```

```python
data_map = []
path="/content/drive/MyDrive/Major Project/Major Project/Sample"
for sub_dir_path in glob.glob(path):
    
    try:
        for filename in os.listdir(sub_dir_path):
          n_path=path+"/"+filename
          print(n_path)
          for f in os.listdir(n_path):
            image_path = n_path + '/' + f
            print(image_path)
            data_map.append(image_path)     
    except Exception as e:
        print(e)
```
